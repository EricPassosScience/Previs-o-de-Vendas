# Previsão de Vendas - Projeto Ciência de Dados

Problema = 

Uma empresa de ecomerce gostaria de saber qual a probalidade de venda nos próximos meses, levando-se em consideração o padrão de invenstimentos em Marketing utilizado até então, através de três grandes veículos informação: TV, Jornal e Rádio. Bem como, qual dos referidos veículos trouxe melhores e piores resultados. 

Observações = 

Os dados de TV, Jornal e Rádio estão no formato de milhares de reais, enquanto os dados das vendas estão em milhões.

1 - Importando as bibliotecas e a base de dados

```python
#Importando as bibliotecas

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
tabela = pd.read_csv("advertising.csv")
display(tabela)
```

2 - Explorando os dados - visualizando: 

- a distribuição dos dados;

- a correlação entre os itens.

```python
sns.heatmap(tabela.corr(), annot=True, cmap="YlGnBu")
plt.show
```

3 - Preparandos os dados para treinar o modelo: dados de treino e dados de teste

```python
y = tabela["Vendas"]
x = tabela.drop("Vendas", axis=1)

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

4 - Em se tratando de um problema de regressão, dois modelos serão analisados:
- Regressão Linear
- Random Forest/Árvore de Decisão

```python
modelo_regressaolinear = LinearRegression()
modelo_arvorededecisao = RandomForestRegressor()

#Treinado o modelo
modelo_regressaolinear.fit(x_treino, y_treino)
modelo_arvorededecisao.fit(x_treino, y_treino)
```

5 - Testando e avaliando o modelo com R²

```python
#previsões
previsao_regressaolinear = modelo_regressaolinear.predict(x_teste)
previsao_arvoredecisao = modelo_arvorededecisao.predict(x_teste)

#Comparação dos modelos 
print(metrics.r2_score(y_teste, previsao_regressaolinear))
print(metrics.r2_score(y_teste, previsao_arvoredecisao))
```

```python
tabela_grafico = pd.DataFrame()

tabela_grafico["y_teste"] = y_teste
tabela_grafico["Previsao Arvore de Decisao"] = previsao_arvoredecisao
tabela_grafico["Previsao Regressao Linear"] = previsao_regressaolinear

plt.figure(figsize=(15,7))
sns.lineplot(data=tabela_grafico)
plt.show()
```

6 - Fazendo a previsão 

```python
novos_dados = pd.read_csv("paraprever.csv")
display(novos_dados)

previsao = modelo_arvorededecisao.predict(novos_dados)
print(previsao)
```

7 - Relação das variaveis com as vendas - Qual o impacto dos meios de comunicação utilizados pelo Marketing nas vendas? 

```python
sns.barplot(x=x_treino.columns, y=modelo_arvorededecisao.feature_importances_)
plt.show()
```

8 - Conclusões

Conclui-se que, dentro dos meios de comunicação escolhidos pelo Marketing, os de maior retorno são os anúncios via TV, seguido pelos via Rádio. 

Além disso, verificou-se que o retorno dos investimos em anúncios via Jornal são insignificantes e, portanto, devem ser descontinuados.
